# Reading Video and Writing Frames

```python
name = "1_Austria_2020"
```

```python
video_path = "../Data/MOV_Files/" + name + ".mov"
saving_dir = "../Data/Frames/" + name + "/"

if os.path.exists(saving_dir):
    print("")
else:
    os.makedirs(saving_dir)
```

```python
def generate_frames(filepath, saving_directory, frame_spacing=10):
    vidcap = cv2.VideoCapture(filepath)

    success, image = vidcap.read()
    count = 0
    
    saving_directory = saving_directory + "/full_frames/"
    
    if os.path.exists(saving_directory):
        print("Directory Exists. Proceeding.")
        shutil.rmtree(saving_directory)
    os.makedirs(saving_directory)
    
    while success:
        success, image = vidcap.read()
        if count%frame_spacing == 0:
            cv2.imwrite(os.path.join(saving_directory, "frame{}.jpg".format(count)), image)
        count += 1
    vidcap.release(); 
```

```python
generate_frames(video_path, saving_dir)
```

# Sub-Frames

```python
def create_subframes(name, coordinates):
    full_frames = natsorted(glob.glob("../Data/Frames/" + name + "/full_frames/*.jpg"))
    
    for coord in coordinates.keys():
        coord_dir = "../Data/Frames/"+ name + "/" + coord + "/"
        if os.path.exists(coord_dir):
            continue
        else:
            os.makedirs("../Data/Frames/"+ name + "/" + coord + "/")
            
    count = 0 
    for file in full_frames:
        img = Image.open(file)
        img = PIL.ImageChops.invert(img)
        for coord in coordinates.keys():
            cropped_img = img.crop(coordinates[coord])
            saving_path = "../Data/Frames/" + name + "/" + coord + "/" + coord + "_{}.png".format(count)
            cropped_img.save(saving_path)

        count = count + 1
```

```python
create_subframes(name, coordinates)
```
